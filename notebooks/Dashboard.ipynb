```python
# ===============================
# 1. Importar bibliotecas
# ===============================
import pandas as pd
```

```python
# ===============================
# 2. Carregar dataset (corrigido)
# ===============================
df = pd.read_csv("Zara_Sales_Analysis.csv", sep=";")

print("Dataset carregado com sucesso!")

```

```python
# ===============================
# 3. Selecionar colunas principais
# ===============================
df_filtrado = df[['Product ID', 'name', 'description', 'Product Category',
                  'Sales Volume', 'price', 'currency', 'section']].head(10)

df_filtrado
```

```python
# ===============================
# 4. Criar coluna de Receita
# ===============================
df_filtrado['Revenue'] = df_filtrado['Sales Volume'] * df_filtrado['price']
df_filtrado
```

```python
# ===============================
# 5. Estatsticas bsicas
# ===============================
print("Total de unidades vendidas:", df_filtrado['Sales Volume'].sum())
print("Receita total:", df_filtrado['Revenue'].sum(), df_filtrado['currency'].iloc[0])
```
